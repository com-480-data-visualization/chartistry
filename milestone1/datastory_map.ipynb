# Interactive map (category, engagement, publication time)

The standalone HTML is built by **`map_ui_export.py`** (Plotly.js + simple controls, no `pip install plotly` needed to view the file).

- **Category** — pick a category, then each country’s **share** of that category’s trending rows (percentages sum to 100% across the ten dataset countries). Sparse categories are flagged in the subtitle.
- **Engagement** — **median** like rate, dislike rate, comment rate, or like/dislike ratio over **all** trending rows in that country.
- **Publication time (UTC)** — scrub hours **0–23**; the map shows each country’s **share of publishes** in that hour (within-country distribution).

**Outputs:** `images/datastory_interactive_map_controls.html` (primary UI) and `images/datastory_interactive_map.html` (same payload, shorter filename). Open either in a browser.

Only the **ten** Kaggle countries are coloured; hover uses full country names.
